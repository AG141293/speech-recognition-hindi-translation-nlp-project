```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

from deep_translator import GoogleTranslator
```

```python
# Load your audio file
audio_file = "/content/diarization_sample_0001.wav"
with sr.AudioFile(audio_file) as source:
    audio_data = recognizer.record(source)
```

```python
try:
    text = recognizer.recognize_google(audio_data)
    print("English Transcription:", text)

    # Translate to Hindi using deep-translator
    hindi_text = GoogleTranslator(source='en', target='hi').translate(text)
    print("Hindi Translation:", hindi_text)

    # Save Hindi text to a file
    with open("transcription_hindi.txt", "w", encoding="utf-8") as file:
        file.write(hindi_text)

except sr.UnknownValueError:
    print("Google Speech Recognition could not understand the audio.")
except sr.RequestError as e:
    print(f"Could not request results from Google Speech Recognition service; {e}")
```
